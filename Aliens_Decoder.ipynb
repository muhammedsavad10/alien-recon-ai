```python
import torch, torchvision
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights

# 1. LOAD THE BRAIN (CNN)
# We use the built-in weights so we don't need to download external JSON labels
weights = EfficientNet_B0_Weights.DEFAULT
model = efficientnet_b0(weights=weights)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

# Get the Earth names directly from the brain's internal dictionary
earth_labels = weights.meta["categories"]

# 2. THE ALIEN DICTIONARY
alien_dictionary = {
    "golden retriever": "Hyperactive Gold-Furred Happiness-Emitter",
    "beagle": "Long-Eared Ground-Sniffer",
    "tabby": "Small Arrogant Predator",
    "bee": "Yellow-Striped Flying Needle-Unit",
    "school bus": "Yellow Metal Box for Young Earthlings",
    "airliner": "Sky-Piercing Metal Tube",
    "broccoli": "Miniature Green Tree (Suspected Toxic)",
    "espresso": "Black Liquid Energy for Organic Engines",
    "pomeranian": "Sentient Cotton Ball",
    "mountain bike": "Two-Wheeled Human Propulsion Device"
}

# 3. THE FIXED SCAN FUNCTION
def xeno_scan_dynamic(image_path):
    try:
        # (Standard Scan Logic - The 'Alien Eye')
        img = Image.open(image_path).convert('RGB')
        preprocess = weights.transforms()
        img_t = preprocess(img).unsqueeze(0).to(device)

        with torch.no_grad():
            output = model(img_t)
            probabilities = torch.nn.functional.softmax(output[0], dim=0)

        prob, index = torch.max(probabilities, 0)
        earth_name = earth_labels[index.item()]
        conf = prob.item() * 100

        # --- DYNAMIC STRATEGIC ADVICE (The Fix) ---
        if "parrot" in earth_name or "lorikeet" in earth_name or "macaw" in earth_name:
            conclusion = "Subject is a feathered vocal mimic."
            advice = "Capture for language decryption. High priority."

        elif "lab coat" in earth_name or "suit" in earth_name:
            conclusion = "Subject is a protective covering for Earthlings."
            advice = "Shred and ignore. Proceed to the next facility."

        elif "dog" in earth_name or "wolf" in earth_name:
            conclusion = "Subject is a pack-hunting quadruped."
            advice = "Exercise caution. Do not allow it to 'fetch'."

        else:
            conclusion = f"Subject identified as {earth_name}."
            advice = "No specific tactical data. Monitor from afar."

        # Display
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"XENO-SCAN: {earth_name.upper()}")
        plt.show()

        print(f"--- üõ∏ XENO-SCANNER INTELLIGENCE REPORT üõ∏ ---")
        print(f"EARTH NAME: {earth_name.upper()} ({conf:.2f}%)")
        print(f"CONCLUSION: {conclusion}")
        print(f"STRATEGIC ADVICE: {advice}")
        print("-" * 45)

    except Exception as e:
        print(f"‚ùå SCANNER ERROR: {e}")

# Try scanning your Parrot again and see if the advice changes!
# xeno_scan_dynamic('parrot.jpg')
# Run it again with your parrot photo!
# xeno_scan_advanced('parrot.jpg')

# TEST IT HERE:
# xeno_scan('parrot.jpg')

# 4. HOW TO TEST
# 1. Upload your photo (e.g., 'cat.jpg') to the folder icon on the left
# 2. Run this line:
# xeno_scan('cat.jpg')
```

```python
import torch
import torchvision
import gradio as gr
from PIL import Image
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights

# 1. SETUP THE ELITE BRAIN
weights = EfficientNet_B0_Weights.DEFAULT
model = efficientnet_b0(weights=weights)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device).eval()
earth_labels = weights.meta["categories"]

# 2. THE ADVANCED ALIEN STYLING
alien_css = """
body { background-color: #050f05; }
.gradio-container { background-color: #050f05 !important; color: #00ff41 !important; border: 1px solid #00ff41; }
#title { text-shadow: 0px 0px 15px #00ff41; font-family: 'Courier New'; }
.stat-box { border: 1px solid #00ff41; padding: 10px; margin: 5px; background: rgba(0, 255, 65, 0.1); }
"""

def xeno_advanced_scan(input_img):
    if input_img is None: return None, "‚ö†Ô∏è SCANNER EMPTY"

    # Preprocess
    preprocess = weights.transforms()
    img_t = preprocess(input_img).unsqueeze(0).to(device)

    # Topic 4: CNN Inference
    with torch.no_grad():
        output = model(img_t)
        probs = torch.nn.functional.softmax(output[0], dim=0)

    # Topic 2 & 5: Rarity Logic (Outlier Detection)
    # We look at the TOP 2 guesses to see how 'confused' the AI is
    top_probs, top_idxs = torch.topk(probs, 2)

    primary_name = earth_labels[top_idxs[0].item()]
    conf = top_probs[0].item() * 100

    # RARITY CALCULATION:
    # If the gap between 1st and 2nd guess is small, the specimen is UNIQUE/RARE.
    # If the overall confidence is low, it's an ANOMALY.
    confidence_gap = (top_probs[0] - top_probs[1]).item()

    if conf < 30 or confidence_gap < 0.1:
        rarity = "LEGENDARY ANOMALY"
        rarity_color = "#ff00ff" # Neon Purple
        action = "CRITICAL: Capture immediately. High Genetic Value."
    elif conf < 60:
        rarity = "RARE SPECIMEN"
        rarity_color = "#00ffff" # Cyan
        action = "Observation required. Unique features detected."
    else:
        rarity = "COMMON EARTHLING"
        rarity_color = "#00ff41" # Matrix Green
        action = "Standard specimen. No immediate value."

    # Build the HTML Dashboard
    report_html = f"""
    <div style="font-family: 'Courier New', monospace; padding: 15px;">
        <h2 style="color: #00ff41; border-bottom: 2px solid #00ff41;">üì° XENO-SCAN COMPLETE</h2>
        <div class="stat-box"><b>EARTH ID:</b> {primary_name.upper()}</div>
        <div class="stat-box"><b>MATCH STRENGTH:</b> {conf:.2f}%</div>
        <div class="stat-box" style="background: {rarity_color}33; border-color: {rarity_color}; color: {rarity_color};">
            <b>SCARCITY STATUS:</b> {rarity}
        </div>
        <hr style="border-color: #004400;">
        <p style="color: #ffffff;"><b>STRATEGIC ACTION:</b> {action}</p>
        <p style="font-size: 10px; color: #006600;">ALGORITHM: EfficientNet-B0 + Confidence Gap Outlier Analysis</p>
    </div>
    """
    return input_img, report_html

# 3. LAUNCH THE COMMAND CENTER
with gr.Blocks(css=alien_css) as demo:
    gr.Markdown("# üõ∏ XENO-COMMAND: ELITE RECON DASHBOARD", elem_id="title")
    with gr.Row():
        with gr.Column():
            input_view = gr.Image(type="pil", label="BIOLOGICAL_INPUT.RAW")
            btn = gr.Button("üëΩ INITIALIZE DEEP SCAN", variant="primary")
        with gr.Column():
            output_view = gr.Image(label="ANALYZED_SPECIMEN.IMG")
            output_html = gr.HTML()

    btn.click(fn=xeno_advanced_scan, inputs=input_view, outputs=[output_view, output_html])

demo.launch(share=True)
```
